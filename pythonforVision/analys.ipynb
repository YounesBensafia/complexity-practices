```python
from matplotlib import pyplot as plt
```

```python
import csv
```

```python
# Initialize lists for each column
valeur = []
N = []
N_div_2 = []
racine_de_N = []

# Open the CSV file
with open('../TP1/MyFile.csv', 'r') as file:
    reader = csv.reader(file)
    
    # Skip the header row
    next(reader)
    
    # Read each row and append values to the appropriate list
    for row in reader:
        valeur.append(float(row[0]))
        N.append(float(row[1]))
        N_div_2.append(float(row[2]))
        racine_de_N.append(float(row[3]))

# Now each list contains data from its respective column
print("Valeur:", valeur)
print("N:", N)
print("N/2:", N_div_2)
print("Racine de N:", racine_de_N)
```

```python
linwidth=1
x=valeur
y=N
z=N_div_2
r=racine_de_N
plt.figure(figsize=(100, 10))
plt.plot(x,y,linewidth=linwidth)
plt.plot(x,z,linewidth=linwidth)
plt.plot(x,r,linewidth=linwidth)
plt.legend(['N','N/2','Sqrt(N)'])
plt.show()
```
